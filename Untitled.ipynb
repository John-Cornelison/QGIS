```python
import re
from shapely.geometry import Polygon

def parse_geometry(geom_str):
    # Remove 'list', 'c', and parentheses
    cleaned_str = re.sub(r'list|c|\(|\)', '', geom_str)
    # Split the string into numbers and group them in pairs
    numbers = [float(num) for num in cleaned_str.split(',')]
    # Group the numbers into pairs (assuming they are coordinates)
    coordinates = list(zip(numbers[::2], numbers[1::2]))
    return Polygon(coordinates)

vet_clinic_by_county['geometry'] = vet_clinic_by_county['geometry'].apply(parse_geometry)
```

```python
import geopandas as gpd

gdf = gpd.GeoDataFrame(vet_clinic_by_county, geometry='geometry')
```

```python
gdf.set_crs(epsg=4326, inplace=True)
```

```python
gdf.to_file('vet_clinic_hours_by_county_shapefile.geojson', driver='GeoJSON')
```
